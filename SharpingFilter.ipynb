```python
check_img = cv2.imread('Lenna.png')
plt.figure(figsize=(3,3))
plt.imshow(check_img[:, :, ::-1])
```

**Running High pass filter on Image**


>  **sharpening filter**



```python
#Set Kernal First
kernel = np.array([ [0,-1,0],
                    [-1,5,-1],
                    [0,-1,0]  ] , dtype="int")
```

```python
final_image = cv2.filter2D(check_img, -1, kernel)
 #plot figure
plt.figure(figsize=[6,6])
#create sub plot for original Image
plt.subplot(121);
plt.imshow(check_img[:,:,::-1]);
plt.title("Original Image",fontdict={'fontsize': 15});plt.axis('off');
#create sub plot for Filtered Image
plt.subplot(122);
plt.imshow(final_image[:,:,::-1]);
plt.title("Sharp Image",fontdict={'fontsize': 15});plt.axis('off');
```

# **Laplacian Implementation**

To mitigate the influence of noise in an image, a recommended approach is to first reduce noise by applying a Gaussian blur. Subsequently, convert the image to grayscale. Following this, apply a Laplacian operator to the grayscale image and visualize the resulting image output.

```python
RGB_img = cv2.cvtColor(check_img, cv2.COLOR_BGR2RGB)
```

```python
src = cv2.GaussianBlur(RGB_img, (3, 3), 0)
# converting to gray scale
gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
edges = cv2.Laplacian(gray, -1, ksize=3, scale=1,delta=0,borderType=cv2.BORDER_DEFAULT)
plt.imshow(edges)
```
