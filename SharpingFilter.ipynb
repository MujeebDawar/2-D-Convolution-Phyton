```python
check_img = cv2.imread('Lenna.png')
plt.figure(figsize=(3,3))
plt.imshow(check_img[:, :, ::-1])
```

**Running High pass filter on Image**


>  **sharpening filter**



```python
#Set Kernal First
kernel = np.array([ [0,-1,0],
                    [-1,5,-1],
                    [0,-1,0]  ] , dtype="int")
```

```python
final_image = cv2.filter2D(check_img, -1, kernel)
 #plot figure
plt.figure(figsize=[6,6])
#create sub plot for original Image
plt.subplot(121);
plt.imshow(check_img[:,:,::-1]);
plt.title("Original Image",fontdict={'fontsize': 15});plt.axis('off');
#create sub plot for Filtered Image
plt.subplot(122);
plt.imshow(final_image[:,:,::-1]);
plt.title("Sharp Image",fontdict={'fontsize': 15});plt.axis('off');
```
