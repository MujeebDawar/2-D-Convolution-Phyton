```python
check_img = cv2.imread('Lenna.png')
plt.figure(figsize=(3,3))
plt.imshow(check_img[:, :, ::-1])
```

```python
kernelSize =(3, 3)
final_image = cv2.GaussianBlur(check_img,kernelSize,cv2.BORDER_DEFAULT)
 #plot figure
plt.figure(figsize=[6,6])
#create sub plot for original Image
plt.subplot(121);
plt.imshow(check_img[:,:,::-1]);
plt.title("Original Image",fontdict={'fontsize': 15});plt.axis('off');
#create sub plot for Filtered Image
plt.subplot(122);
plt.imshow(final_image[:,:,::-1]);
plt.title("Smooth Image",fontdict={'fontsize': 15});plt.axis('off');
```
