```python
equ = cv2.equalizeHist(check_img,0)

plt.figure(figsize=[7,7])
plt.subplot(121);plt.imshow(check_img,cmap='gray');plt.title("Original");plt.axis('off');
plt.subplot(122);plt.imshow(equ,cmap='gray');plt.title("Equalized Image");plt.axis('off');
```
